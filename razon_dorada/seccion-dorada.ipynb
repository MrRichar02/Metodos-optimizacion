```python
f = lambda x: 2*np.sin(x) - x**2/10
```

```python
def seccion_dorada(f, xu, xl, error):
    interaciones = []
    PHI = (1+np.sqrt(5))/2
    CONSTANT = 1/PHI
    d = CONSTANT*(xu-xl)
    x1 = xl+d
    x2 = xu-d
    error_actual = abs(x1-x2)
    fx1 = f(x1)
    fx2 = f(x2)
    decision = fx2 > fx1
    interaciones.append([xl, xu, d, x1, x2, fx1, fx2, "f(x2)>f(x1)" if decision else "f(x2)<f(x1)", error_actual])
    while(error_actual > error):
        if decision:
            xu = x1
        else:
            xl = x2
        d = CONSTANT*(xu-xl)
        x1 = xl+d
        x2 = xu-d
        error_actual = abs(x1-x2)
        fx1 = f(x1)
        fx2 = f(x2)
        decision = fx2 > fx1
        interaciones.append([xl, xu, d, x1, x2, fx1, fx2, "f(x2)>f(x1)" if decision else "f(x2)<f(x1)", error_actual])
    df = pd.DataFrame(interaciones, columns=["XL", "XU", "d", "X1", "X2", "F(X1)", "F(X2)", "Evaluacion", "Error"])
    return df
        

```

```python
seccion_dorada(f, 4, 0, 0.001)
```
