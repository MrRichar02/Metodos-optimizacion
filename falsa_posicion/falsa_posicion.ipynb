```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
function = lambda x: 3*x**2 - 120*x +100
```

```python
def falsa_posicion(funcion, xu, xl, error):
    if funcion(xu)*funcion(xl) >= 0:
        print("La función no cambia de signo en el intervalo dado.")
        return None
    df = pd.DataFrame(columns=['xu', 'xl', 'xr', 'f(xr)', 'Error'])
    xr = xu-(funcion(xu)*(xl-xu))/(funcion(xl)-funcion(xu))
    decicion = funcion(xr)*funcion(xl)
    while(xu-xr > error):
        if decicion < 0:
            xu = xr
        elif decicion > 0:
            xl = xr
        else:
            print("xr es la raíz exacta.")
            return xr
        xr = xu-(funcion(xu)*(xl-xu))/(funcion(xl)-funcion(xu))
        decicion = funcion(xr)*funcion(xl)
    return xr
```

```python
falsa_posicion(function, 4, 0, 0.01)
```

```python
falsa_posicion(function, 41, 15, 0.01)
```
