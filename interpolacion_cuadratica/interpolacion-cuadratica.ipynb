# Interpolación cuadrática

```python
def interpolacion_cuadratica(x0, x1, x2, f, error):
    iteraciones = []
    while True:
        fx0, fx1, fx2 = f(x0), f(x1), f(x2)
        numerador = (fx0*(x1**2 - x2**2) + fx1*(x2**2 - x0**2) + fx2*(x0**2 - x1**2))
        denominador = (2*fx0*(x1 - x2) + 2*fx1*(x2 - x0) + 2*fx2*(x0 - x1))

        # Verificar si el denominador es cero antes de realizar la división
        if denominador == 0:
            raise ValueError("Denominador es cero, no se puede continuar con la interpolación.")

        x3 = numerador/denominador
        fx3 = f(x3)
        error_actual = abs(x3-x1)

        # Guardar la iteración actual
        iteraciones.append([x0, x1, x2, x3, fx0, fx1, fx2, fx3, error_actual])

        if (x3 >= x1 and fx3 > fx1):
            x0 = x1
            x1 = x3
        elif (x3 >= x1 and fx3 < fx1):
            x2 = x3
        elif (x3 < x1 and fx3 < fx1):
            x0 = x3
        elif (x3 < x1 and fx3 > fx1):
            x2 = x1
            x1 = x3
        
        if error_actual <= error:
            break
    return pd.DataFrame(iteraciones, columns=['x0', 'x1', 'x2', 'x3', 'f(x0)', 'f(x1)', 'f(x2)', 'f(x3)', 'Error'])
```

```python
f = lambda x: 2*np.sin(x) - x**2/10
x0 = 0
x1 = 1
x2 = 4
```

```python
interpolacion_cuadratica(x0, x1, x2, f, 0.01)
```
